```python
import adgeek_permission as permission
import facebook_business.adobjects.campaign as facebook_business_campaign
import facebook_business.adobjects.ad as facebook_business_ad
import facebook_business.adobjects.adcreative as facebook_business_adCreative
import facebook_business.adobjects.pagepost as facebook_business_pagepost
```

```python
account_id = 267083990698486
campaign_id = 23843098840440451
permission.init_facebook_api(account_id)

def get_ads_by_campaign(campaign_id):
    this_campaign = facebook_business_campaign.Campaign(campaign_id)
    this_campaign_ads = this_campaign.get_ads()
    ad_id_list = []
    for this_ad in this_campaign_ads:
        ad_id = int(this_ad.get('id'))
        ad_id_list.append(ad_id)
    return ad_id_list
```

```python
ad_id_list = get_ads_by_campaign(campaign_id)
ad_id_list
```

```python
def get_creative_by_ad(ad_id):
    this_ad = facebook_business_ad.Ad(ad_id)
    this_creatives = this_ad.get_ad_creatives()
    creative_id_list = []
    for this_creative in this_creatives:
        creative_id = int(this_creative.get('id'))
        creative_id_list.append(creative_id)
    return creative_id_list
```

```python
ad_id = 23843708427040451
creative_id_list = get_creative_by_ad(ad_id)
creative_id_list
```

```python

def get_post_by_creative(creative_id):
    this_creative = facebook_business_adCreative.AdCreative(creative_id)
    this_object_story_id = this_creative.api_get(fields=['object_story_id'])
    object_story_id = this_object_story_id.get('object_story_id')
    return object_story_id
    
```

```python
creative_id = 23843708446390451
post_id = get_post_by_creative(creative_id)
```

```python
def get_call_to_action_link_by_post(post_id):
    this_post = facebook_business_pagepost.PagePost(post_id)
    this_post_info = this_post.api_get(fields=['call_to_action'])
    call_to_action = this_post_info.get('call_to_action')
    if call_to_action:
        call_to_action_value = call_to_action.get('value')
        if call_to_action_value:
            call_to_action_link = call_to_action_value.get('link')
            if call_to_action_link:
                return call_to_action_link
```

```python
post_id = '820612794778213_1252928054880016'
call_to_action_link = get_call_to_action_link_by_post(post_id)
call_to_action_link
```

```python
##test total campaign
ad_id_list = get_ads_by_campaign(campaign_id)
for ad_id in ad_id_list:
    print('ad_id:',ad_id)
    creative_id_list = get_creative_by_ad(ad_id)
    for creative_id in creative_id_list:
        print('creative_id:',creative_id)
        post_id = get_post_by_creative(creative_id)
        print('post_id:',post_id)
        call_to_action_link = get_call_to_action_link_by_post(post_id)
        if call_to_action_link:
            print('call_to_action_link:', call_to_action_link)
    print('--')
```
