```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
perform = pd.read_csv('../data/act_1587964461298459_ad.csv')
activity = pd.read_csv('../data/act_1587964461298459_activity_log.csv')
```

```python
perform = perform.drop(['ad_name', 'Unnamed: 14'], axis=1)
```

```python
perform['date_start'] = pd.to_datetime(perform['date_start'])
perform = perform.sort_values(['campaign_id', 'adset_id', 'ad_id', 'date_start', 'hourly_stats_aggregated_by_audience_time_zone']).reset_index(drop=True)
```

```python
activity['event_time'] = pd.to_datetime(activity['event_time'])
activity['event_date'] = activity['event_time'].dt.date
activity['hour'] = activity['event_time'].dt.hour
```

```python
activity = activity.sort_values(['object_id','event_date','hour']).reset_index(drop=True)
```

```python
for i in range(len(activity)):
    if activity['new_bid'].iloc[i] < 0:
        print(i)
```

```python
adid_list = perform['adset_id'].unique()
```

```python

for adset in adid_list:
    fig=plt.figure(figsize=(15,3))
    plt.plot(activity['event_date'][activity.object_id==adset], activity['new_bid'][activity.object_id==adset], label=adset)
#     plt.plot(perform['date_start'][perform.adset_id==adset], perform['clicks'][perform.adset_id==adset], label=adset)
    plt.grid(True)
    plt.legend()
    plt.show()
#     plt.legend(loc='upper left',bbox_to_anchor=(1,1.04))
```
